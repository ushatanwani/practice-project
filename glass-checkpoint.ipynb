```python
dfnn=df.drop([105,106,107,112,163,171,172,184,201],axis=0)
```

```python
dfnn['Ba:Barium']=np.cbrt(dfnn['Ba:Barium'])
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
```

```python
x=dfnn.iloc[:,0:10]
y=dfnn['Type of glass']
```

```python
x.shape
```

```python
y.unique()
```

```python
y.shape
```

```python
y.replace(1,0,inplace=True)
y.replace(2,0,inplace=True)
y.replace(3,0,inplace=True)
y.replace(5,1,inplace=True)
y.replace(6,1,inplace=True)
y.replace(7,1,inplace=True)
```

```python
y.unique()
```

```python
x.head()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif['VIF Values']=[variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
vif['Features']=x.columns
vif
```

```python
from imblearn import under_sampling, over_sampling
from imblearn.over_sampling import SMOTE

sm=SMOTE()
trainx,trainy=sm.fit_resample(x,y)
trainy.value_counts()
trainx.value_counts()
```

```python
y.value_counts()
```

```python
trainy.shape
```

```python
trainx.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report,confusion_matrix,roc_curve,accuracy_score
```

```python
maxaccu=0
maxrs=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(trainx,trainy,test_size=0.30,random_state=i)
    rfr=RandomForestClassifier()
    rfr.fit(x_train,y_train)
    pred=rfr.predict(x_test)
    acc=accuracy_score(y_test,pred)
    if acc>maxaccu:
        maxaccu=acc
        maxrs=i
print('Best accuracy is',maxaccu ,"at random_state",maxrs)
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
x_train,x_test,y_train,y_test=train_test_split(trainx,trainy,test_size=.30,random_state=maxrs)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import GradientBoostingClassifier,AdaBoostClassifier,BaggingClassifier
from sklearn.metrics import classification_report,confusion_matrix,roc_curve,accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train,y_train)
predsvc=svc.predict(x_test)
print(accuracy_score(y_test,predsvc))
print(confusion_matrix(y_test,predsvc))
print(classification_report(y_test,predsvc))
```

```python
gb=GradientBoostingClassifier()
gb.fit(x_train,y_train)
predgb=gb.predict(x_test)
print(accuracy_score(y_test,predgb))
print(confusion_matrix(y_test,predgb))
print(classification_report(y_test,predgb))
```

```python
ad=AdaBoostClassifier()
ad.fit(x_train,y_train)
predad=ad.predict(x_test)
print(accuracy_score(y_test,predad))
print(confusion_matrix(y_test,predad))
print(classification_report(y_test,predad))
```

```python
#hyperparameter tuning
```

```python
from sklearn.model_selection import GridSearchCV 
parameters = {'algorithm': ['SAMME', 'SAMME.R'],"random_state":[10,50,1000],'learning_rate': [0.0001, 0.001, 0.01, 0.1, 1.0],"n_estimators":[50,100,200,300]}
ad=AdaBoostClassifier()
gcv=GridSearchCV(ad,parameters,cv=5)
gcv.fit(x_train,y_train)
gcv.best_params_
```

```python
final_model=AdaBoostClassifier(algorithm='SAMME',learning_rate=0.0001,n_estimators=50,random_state=10)
final_model.fit(x_train,y_train)
predad=final_model.predict(x_test)
print(accuracy_score(y_test,predad))
print(confusion_matrix(y_test,predad))
print(classification_report(y_test,predad))
```

```python
from sklearn import metrics
from sklearn import model_selection
from sklearn.metrics import plot_roc_curve 
```

```python
plot_roc_curve(final_model,x_test,y_test)
plt.title('roc for the best model')
plt.show()
```

```python
from sklearn.metrics import roc_auc_score,auc
from sklearn.metrics import roc_curve
```

```python
y_pred_prob = final_model.predict_proba(x_test)[:,1]
fpr,tpr,thresholds = roc_curve(y_test,y_pred_prob)
roc_auc = auc(fpr,tpr)
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr,label='ExtraTreesClassifier(area=%0.2f)'%roc_auc)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
auc_score= roc_auc_score(y_test,final_model.predict(x_test))
print('Auc score:','/n',auc_score)
```
